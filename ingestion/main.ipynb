```python
import subprocess
import sys
```

```python
PIPELINE = [
    "../etl/landing2raw.ipynb",
    "../etl/raw2cleansed.ipynb",
    #"../etl/cleansed2curated.ipynb",
]
```

```python
def run_step(script: str) -> None:
    print(f"\n{'='*50}\n▶ Iniciando: {script}\n{'='*50}")
    result = subprocess.run(
        ["jupyter", "nbconvert", "--to", "notebook", "--execute", "--inplace", script],
        check=False,
    )
    if result.returncode != 0:
        print(f"\n✗ Falha em '{script}' (exit code {result.returncode}). Pipeline interrompido.")
        sys.exit(result.returncode)
    print(f"✔ Concluído: {script}")
```

```python
for script in PIPELINE:
    run_step(script)

print("\n✔ Pipeline concluído com sucesso.")
```
